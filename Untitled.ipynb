```python
#recuperation des produit ach√©t√© du dataset sans doublon
liste= donnee['StockCode'].unique() 
#Conversion du dataset en liste
new_donnes = donnee.values.tolist()
def somme_quantity_par_produit(liste,new_donnes):
    tab=[]
    tab2=[]
    #Sommation des quantit√©s de chaque produit ach√©t√© du dataset
    for i in liste:
        tab1=[]  
        for j in new_donnes:
            if j[1]==i:
                tab1.append(j[3])
        tab.append(sum(tab1))
        tab2.append(i)
    return tab, tab2
```

```python
#conversion des deux listes obtenus en dataframe
tab, tab2=somme_quantity_par_produit(liste,new_donnes)
data = pd.DataFrame((zip(tab2,tab)), columns=['Produit','Quantity_toal']) 
#Triallage du dataframe
data=data.sort_values(by = 'Quantity_toal', ascending = False)
data
```

```python
fig, axes = plt.subplots(1,1, figsize=(12,5))
axes.scatter(data=data,x='Produit',y='Quantity_toal')
axes.set_xlabel("Produit")
axes.set_ylabel("Quantity toal")
plt.tight_layout()
plt.show()
```
